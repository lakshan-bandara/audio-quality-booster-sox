```python
#@title ⚙️ Step 1: Install Dependencies (Run Once)
#@markdown Click **▶ Run** to install SoX and required libraries.

import subprocess, sys, os

print("📦 Installing SoX and required packages...")
subprocess.run(['apt-get', 'install', '-y', 'sox', 'libsox-fmt-all'], 
               capture_output=True, text=True)
subprocess.run([sys.executable, '-m', 'pip', 'install', 'pydub', 'ipywidgets'], 
               capture_output=True, text=True)

print("✅ Installation complete! Proceed to Step 2.")
```

```python
#@title 📂 Step 2: Upload Your Audio File
#@markdown Upload your noisy audio file (WAV format recommended).

from google.colab import files
import os, shutil

print("📁 Please upload your audio file (WAV preferred):")
uploaded = files.upload()

if uploaded:
    filename = list(uploaded.keys())[0]
    print(f"✅ Uploaded: {filename}")
    print(f"📊 File size: {os.path.getsize(filename) / 1024:.1f} KB")
else:
    print("❌ No file uploaded. Please try again.")
```

```python
#@title 🔍 Step 3: Generate Noise Profile
#@markdown This step analyzes the background noise in your audio.
#@markdown 
#@markdown **Noise Sample Duration (seconds):**
noise_sample_duration = 1.5  #@param {type:"number"}

import subprocess, os

try:
    filename = [f for f in os.listdir('.') if f.endswith(('.wav', '.WAV'))][0]
    print(f"🎵 Processing: {filename}")
except:
    print("❌ No WAV file found. Please complete Step 2 first.")
    raise SystemExit

print(f"📡 Sampling first {noise_sample_duration}s for noise profile...")

# Extract noise sample
result = subprocess.run(
    ['sox', filename, '-n', 'trim', '0', str(noise_sample_duration), 'noiseprof', 'noise.prof'],
    capture_output=True, text=True
)

if os.path.exists('noise.prof'):
    print("✅ Noise profile created successfully!")
    print(f"📄 Profile size: {os.path.getsize('noise.prof')} bytes")
else:
    print(f"❌ Error: {result.stderr}")
```

```python
#@title 🚀 Step 4: Apply Noise Reduction
#@markdown Configure noise reduction settings below:

#@markdown ---
#@markdown **Noise Reduction Sensitivity** (0.1 = light, 0.5 = aggressive):
sensitivity = 0.21  #@param {type:"slider", min:0.1, max:0.5, step:0.01}

#@markdown **Output Filename:**
output_name = "cleaned_audio"  #@param {type:"string"}

#@markdown **Apply Audio Normalization:**
normalize = True  #@param {type:"boolean"}

import subprocess, os

try:
    filename = [f for f in os.listdir('.') if f.endswith(('.wav', '.WAV'))][0]
except:
    print("❌ No WAV file found. Please complete Steps 2 & 3 first.")
    raise SystemExit

output_file = f"{output_name}.wav"

print(f"🎛️ Applying noise reduction (sensitivity: {sensitivity})...")

# Build SoX command
sox_cmd = ['sox', filename, output_file, 'noisered', 'noise.prof', str(sensitivity)]
if normalize:
    sox_cmd += ['norm', '-3']

result = subprocess.run(sox_cmd, capture_output=True, text=True)

if os.path.exists(output_file):
    orig_size = os.path.getsize(filename) / 1024
    clean_size = os.path.getsize(output_file) / 1024
    print(f"✅ Noise reduction complete!")
    print(f"📊 Original: {orig_size:.1f} KB → Cleaned: {clean_size:.1f} KB")
    print(f"💾 Output saved as: {output_file}")
else:
    print(f"❌ Error: {result.stderr}")
```

```python
#@title 🎧 Step 5: Preview & Compare Audio
#@markdown Listen to both original and cleaned audio side by side.

import os
from IPython.display import Audio, display, HTML

try:
    filename = [f for f in os.listdir('.') if f.endswith(('.wav', '.WAV')) and 'cleaned' not in f][0]
    cleaned = [f for f in os.listdir('.') if 'cleaned' in f and f.endswith('.wav')][0]
except:
    print("❌ Audio files not found. Please complete previous steps.")
    raise SystemExit

display(HTML("<h3>🔴 Original (Noisy) Audio:</h3>"))
display(Audio(filename, autoplay=False))

display(HTML("<h3>🟢 Cleaned Audio (Noise Removed):</h3>"))
display(Audio(cleaned, autoplay=False))

print("\n🎉 Compare the two audio files above!")
```

```python
#@title 📥 Step 6: Download Cleaned Audio
#@markdown Download your noise-reduced audio file.

from google.colab import files
import os

try:
    cleaned = [f for f in os.listdir('.') if 'cleaned' in f and f.endswith('.wav')][0]
    print(f"⬇️ Downloading: {cleaned}")
    files.download(cleaned)
    print("✅ Download started!")
except:
    print("❌ Cleaned audio not found. Please complete Steps 2-4 first.")
```
